```python
import sys
import numpy
import pandas as pd
import matplotlib
import scipy as sp
import IPython
import sklearn
```

```python
# load dataset from scikit-learn practice datasets
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
# Familarizing myself with the data
#
# The dataset is a colection of nested arrays.
# The key for the array describe each array's data.
#
print("Keys of iris_dataset: \n{}".format(iris_dataset.keys()))
```

```python
# 
# feature_names = description of the dat in the 'data' key
#
print(format(iris_dataset['feature_names']))
```

```python
#
# data = arrays of various measurements of flowers, in order of 'feature_names'
#
print(format(iris_dataset['data'][:5]))
```

```python
#
# target_names = names of various species of irises identified by botantists
#
print(format(iris_dataset['target_names']))
```

```python
#
# target = species data for all measured irises, coresponds to labels in 'target_names'
#
print(format(iris_dataset['target']))
```

```python
#
# DESCR = detailed describtion of the dataset including references
#
print(format(iris_dataset['DESCR'][:150]))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
```

```python
#
# library of utility function written specifically for O'Reily Intro. to M.L.
#
# import mglearn
#
# only used here for matplotlib ListedColormap, instead lets just define the color map by hand instead of importing a library
#
from matplotlib.colors import ListedColormap
cm3 = ListedColormap(['#0000aa', '#ff2020', '#50ff50'])
```

```python
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=cm3)
```
